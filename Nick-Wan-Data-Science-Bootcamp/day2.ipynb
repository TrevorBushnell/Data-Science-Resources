## Read in the Data and Initialize Packages

```python
import pandas as pd

df = pd.read_csv('data/2023-nwbc-reds.csv')
df.head()
```

## And and Or Statements in Python

* To check if two conditions are true, use an AND statement with the `&` symbol
* To check if either of two conditions are true, use an OR statement with the `|` symbol

* Find the rows where the home team is CIN and it is the top of the inning

```python
df.loc[(df['home_team']=='CIN') & (df['inning_topbot']=='Top')].head()
```

* Create a new column called `is_reds_pitcher` where the home team is CIN and it is the top of the inning OR the away team is CIN and it is the bottom of the inning

```python
df['is_reds_pitcher'] = 0
df.loc[((df['home_team']=='CIN') & (df['inning_topbot']=='Top')) | (df['away_team']=='CIN') & (df['inning_topbot']=='Bot'), 'is_reds_pitcher'] = 1
df.head()
```

## Some Nice Python Shorthand

* If you want to get everything (either all rows or all columns), then use `:`
* If you want to get specific columns, put the column names into a list

EX: Get all the rows from your dataframe with only the pitcher name, pitch name, and whether the pitch was a strike columns

```python
df.loc[:, ['pitcher_name', 'pitch_name', 'is_strike']].head()
```

## Pivot Tables in Python

* We can create pivot tables from our pandas DataFrames using the `groupby()` function
* The result of this is NOT a dataframe... so if you want some aggragation functions just tack them on to the end with normal function notation

```python
df.loc[df['is_reds_pitcher']==1, ['pitcher_name', 'pitch_name', 'is_strike']].groupby(['pitcher_name', 'pitch_name'], as_index=False).mean()
```
