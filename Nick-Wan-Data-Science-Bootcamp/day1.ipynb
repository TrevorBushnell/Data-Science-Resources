## Import Our Dataset

* datasets are manipulated with the `pandas` package as python
  * we usually call it `pd` so that we can not have to type as many characters when we write code lol
* pandas has a data structure called a **data frame** which stores data stored in tables
  * handy because we can have labeled indices which is way handier when dealing with tables!
  * we store dataframes in variables and add `df` to the variable name for naming conventions
* to read in data from a csv file, use the `pd.read_csv()` function
* to show the first few rows of the dataframe, use the `head()` function
  * default 5, but can pass in any number to get that many rows

```python
import pandas as pd

df = pd.read_csv('data/2023-nwbc-reds.csv')
df.head()
```

* You can get the names of all the columns in your dataframe by calling the `columns` attribute

```python
df.columns
```

Since dataframes use labeled indices, we can get a column by writing the column name inside the square brackets:

```python
df['home_score']
```

* We can add two columns and set that equal to a new column really easily!

```python
df['total_score'] = df['home_score'] + df['away_score']
df.head()
```

* We can have conditionals to create unique statements with dataframes in order to create new columns
* to do this, we can start by getting all the unique values in a column using the `unique()` function:

```python
df['description'].unique()
```

* Now we can create this new column of whether an event was a strike using these values
* Btw, the `loc` method allows us to locate exact values in a dataframe
* AND you can autofill a column with a value by creating a new column and setting it equal to a single value that you want inserted everywhere

```python
strike_events = ['called_strike', 'swinging_strike', 'foul', 'swinging_strike_blocked', 'foul_trip', 'foul_bunt', 'missed_bunt', 'bunt_foul_tip']

df['is_strike'] = 0
df.loc[df['description'].isin(strike_events), 'is_strike'] = 1
df.head()
```

## Homework

1. Use the `strikes` column to find where there are 2 strikes and make a new column indicating whether the pitch was thrown in a 2-strike situation. Name this `is_2_strikes`.

```python
df['is_2_strikes'] = 0
df.loc[df['strikes'] == 2, 'is_2_strikes'] = 1

df.head()
```

2. Use the `release_speed` column to find when a pitch is thrown greater than 94 MPH (the column is in MPH already). Name this `is_high_velo`

```python
df['is_high_velo'] = 0
df.loc[df['release_speed'] > 94, 'is_high_velo'] = 1

df.head()
```

3. Use the `description` column and the `attack_zone` column to indicate when a pitch is thrown for a called or swinging stike (not a foul) in the "heart" attack zone (hint: this requires two conditional statements! maybe google or ask for help in the discord)

```python
strike_events = ['called_strike', 'swinging_strike']

# way #1: using &
df['strike_in_heart'] = 0
df.loc[(df['description'].isin(strike_events)) & (df['attack_zone'] == 'heart'), 'strike_in_heart'] = 1
df.head()
```

```python
# write our work to a csv so we can use it later
df.to_csv('./data/2023-nwbc-reds.csv', index=False)
```
